```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = np.random.rand(1000, 1000)
row_sums = data.sum(axis=1)  # Calculate the sum of each row
normalized_data = data / row_sums[:, np.newaxis]  # Normalize each row by dividing by its sum
plt.figure(figsize=(10, 10))
sns.heatmap(normalized_data, cmap='Oranges', cbar=False, square=True)
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.xticks([])  # Remove x-axis ticks
plt.yticks([])  # Remove y-axis ticks
plt.show()
```

```python
import seaborn as sns
from scipy.stats import ttest_ind

# Generate some example data
data = sns.load_dataset('tips')

# Create violin plot
sns.violinplot(x='day', y='total_bill', hue='sex', data=data, split=True)

# Perform t-test between the distributions specified by hue
group1 = data[data['sex'] == 'Male']['total_bill']
group2 = data[data['sex'] == 'Female']['total_bill']
t_statistic, p_value = ttest_ind(group1, group2)

# Display p-value on the plot
plt.text(0.5, 0.8, f'p-value: {p_value:.4f}', ha='center', va='center', transform=plt.gca().transAxes)

# Show the plot
plt.show()
```
